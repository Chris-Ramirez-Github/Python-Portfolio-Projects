# For Loops

```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
    print(number)
```

```python
ice_cream_dict = {'name': 'Chris Ramirez', 'weekly intake': 5, 'favorite ice creams': ['MCC', 'Chocolate']}
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key, '-->',value)
```

# Nested For Loop

```python
flavors = ['Vanilla', 'Chocolate','Cookie Dough']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']
```

```python
for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
```

# While Loop

```python
number = 0

while number < 5:
    print(number)
    number = number + 1 
```

```python
number = 0

while number < 5:
    number += 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')
    
```
