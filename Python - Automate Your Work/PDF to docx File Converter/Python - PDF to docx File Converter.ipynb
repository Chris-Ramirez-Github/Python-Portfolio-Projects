```python
# import pytube library to downlaod the video
from pdf2docx import Converter
import PySimpleGUI as sg 


# Set the theme for GUI
sg.theme('Default1')

# Prompt user to enter countdown time
layout = [
            [sg.Text("Convert PDF to docx\n")],
# Ask to upload PDF file
            [sg.Text("Select a pdf file: "), sg.Input(size=(25,1) , key="-IN-"), sg.FileBrowse()],
            [sg.Text("Select a folder destination: "), sg.Input(size=(25,1), key="-FOLDER-"), sg.FolderBrowse()],
            [sg.Text("Choose a docx file name include .docx: "), sg.InputText(size=(25,1), key="-File-")],
            [sg.Text("\n")],
            [sg.Button('Convert'),sg.Button('Cancel')]
            
]

# Create the Window
window = sg.Window('PDF to docx File Convertor', layout)

while True:
        # Read user input
        event, values = window.read()
        pdf_file = values["-IN-"]
        docx_file = values["-FOLDER-"]
        file_name = docx_file +'/'+ values["-File-"]
       
        if event == sg.WIN_CLOSED or event == 'Cancel':
            break
# Call Countdown() function if user clicks start button                       
        if event == 'Convert':
            sg.popup("Converting pdf to docx file now...")
            cv = Converter(pdf_file)
            cv.convert(file_name)
            cv.close()
            
# Break while loop if window is closed or user clicks cancel
        elif event == sg.WIN_CLOSED or event == 'Cancel':
            break
        else:
            sg.popup_error("Enter valid input")
        
# if user closes window or clicks cancel
        if event == sg.WIN_CLOSED or event == 'Cancel':
            break
        
# Close window   
window.close()  
```

### 1. Select PDF file to convert into docx file

### 2. Select the file destination

### 3. Choose a file name including the .docx extension

### 4. Click Convert button to start the file conversion process

### 5. The new docx file is ready for viewing
