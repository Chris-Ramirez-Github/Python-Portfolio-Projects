```python
import PySimpleGUI as sg

sg.theme('SystemDefault')

layout = [
            # Requesting user input for name, weight and height
            [sg.Text("Calculate Your Body Mass Index\n")],
            [sg.Text("Enter your name:"), sg.InputText()],
            [sg.Text("Enter your weight in pounds:"), sg.InputText()],
            [sg.Text("Enter your height in inches: "), sg.InputText()],
            [sg.Text("\n")],
            [sg.Button('Submit'),sg.Button('Cancel')]
            
]

# Create the Window
window = sg.Window('BMI Calculator', layout, size=(600, 200))

# Event loop to process "events" and get the "values of inputs"
while True:
    event, values = window.read()
    
    name = values[0]
    weight = int(values[1])
    height = int(values[2])
    
    if event == 'Submit':
        BMI = round((weight * 703) / (height * height),1)
        if BMI>0:
            if(BMI<18.5):
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are underwight.")
            elif (BMI<=24.9):
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are normal weight.")
            elif (BMI<29.9):
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are overweight.")
            elif (BMI<34.9):
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are obese.")
            elif (BMI<39.9):
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are severely obese.")
            else:
                sg.popup(name +" your BMI is "+ str(BMI) + ", you are morbidly obese.")
    elif event == sg.WIN_CLOSED or event == 'Cancel': 
        break
    else:
        sg.popup("Enter valid input")
        
# if user closes window or clicks cancel
    if event == sg.WIN_CLOSED or event == 'Cancel': 
        break
    
window.close()
```

## Formula used to calculate BMI
### BMI = (weight in pounds x 703) / (height in inches x height in inches)

## BMI parameters used to calculate are provided by Mercer Health
### Under 18.5	Underweight	Minimal
### 18.5 - 24.9	Normal Weight	Minimal
### 25 - 29.9	Overweight	Increased
### 30 - 34.9	Obese	High
### 35 - 39.9	Severely Obese	Very High
### 40 and over	Morbidly Obese	Extremely High
